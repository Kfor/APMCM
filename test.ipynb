```python
import cv2  
import numpy as np  
import os 
import matplotlib.pyplot as plt


path = r'C:\Users\Kfor\Desktop\亚太\2019 APMCM Problems\2019 APMCM Problems\InnerArea'
os.chdir(path)



firstframe = cv2.imread('0610.bmp',cv2.IMREAD_GRAYSCALE)
frame = cv2.imread('0499.bmp',cv2.IMREAD_GRAYSCALE)


gray=cv2.GaussianBlur(frame,(21,21),0)  



frameDelta = cv2.absdiff(firstframe,gray)  
thresh = cv2.threshold(frameDelta, 25, 255, cv2.THRESH_BINARY)[1]  
thresh = cv2.dilate(thresh, None, iterations=2)  
# cnts= cv2.findContours(thresh.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)  
    
x,y,w,h=cv2.boundingRect(thresh)  
frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2) 
 


plt.subplot(1,2,1)
plt.imshow(thresh,cmap=plt.cm.gray)  
plt.subplot(1,2,2)
plt.imshow(frameDelta,cmap=plt.cm.gray)  
    
```
